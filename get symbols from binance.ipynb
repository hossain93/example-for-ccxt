# with load_markets

```python
import pandas as pd
import ccxt
binance = ccxt.binance({'options': { 'defaultType': 'future' , 'contractType': 'PERPETUAL' , 'quote': 'USDT' }})
#binance.fetchMarkets ()
binance.load_markets()
pairs=pd.DataFrame(binance.load_markets ()).transpose()  #pairs=pd.DataFrame(binance.fetchMarkets ())
pd.set_option('display.max_row', pairs.shape[0]+1)
pairs
pairs.drop(pairs.loc[pairs['quote']!='USDT'].index, inplace=True)
pairs
symbols=pairs['symbol']
pd.DataFrame(symbols)
symbols.reset_index(drop=True, inplace=True)
symbols
```

# with fetchMarkets

```python
import pandas as pd
import ccxt
binance = ccxt.binance({'options': { 'defaultType': 'future' , 'contractType': 'PERPETUAL' , 'quote': 'USDT' }})
#binance.fetchMarkets ()
binance.load_markets()
pairs=pd.DataFrame(binance.fetchMarkets ())
pd.set_option('display.max_row', pairs.shape[0]+1)
pairs
pairs.drop(pairs.loc[pairs['quote']!='USDT'].index, inplace=True)
pairs
symbols=pairs['symbol']
pd.DataFrame(symbols)
symbols.reset_index(drop=True, inplace=True)
symbols
```
